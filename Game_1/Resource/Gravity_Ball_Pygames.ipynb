# Simple Drag and Drop Gravity Ball

```python
import pygame
pygame.init()

xmin = ymin = 0
xmax = ymax = 700

win = pygame.display.set_mode((xmax,ymax))
pygame.display.set_caption("Gravity Ball")

x = 50
y = 50
ratio = 20/3
radius = thickness = 20
velx = 15
vely=0
elastic_coeff = 0.85
g =1# 980*ratio

run = True
time = 30
b = 0
while run:
    pygame.time.delay(time)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    
    
    
    
    
    ytemp = y+vely+g/2
    
    if ymin+2*radius<=ytemp+radius <= ymax:
        vely = vely +g
        y = y+vely+g/2
    elif ymin+2*radius>ytemp+radius:
        vely = - elastic_coeff*vely
        y = ymin+radius
    elif ytemp+radius > ymax:
        vely = - elastic_coeff*vely
        y = ymax-radius    
        
    
    xtemp = x+velx
    
    if xmin+2*radius<=xtemp+radius <= xmax and y !=ymax-radius :
        x = x+velx
    elif xmin+2*radius<=xtemp+radius <= xmax and y==ymax-radius :
        velx = 0.95*velx
        x = x+velx
    elif xmin+2*radius>xtemp+radius:
        velx = - elastic_coeff*velx
        x = xmin+radius
    elif xtemp+radius > xmax:
        velx = - elastic_coeff*velx
        x = xmax-radius
    
    if pygame.mouse.get_pressed()[0]:
        if b == 0:
            x1,y1 =pygame.mouse.get_pos()
            b = 1
        if abs(x1-x)<25 and abs(y1-y)<25:
            x,y = pygame.mouse.get_pos()
            velx = vely = 0
        x1,y1 = pygame.mouse.get_pos()
        
    
    
    win.fill((255,255,255))
    pygame.draw.circle(win, (0,0,0), (x, y), radius, thickness)   
    pygame.display.update() 
    
pygame.quit()
```

# Intractive Gravity Ball

```python
import pygame
pygame.init()

xmin = ymin = 0
xmax = ymax = 500

win = pygame.display.set_mode((xmax,ymax))
pygame.display.set_caption("Gravity Ball")

x = 50
y = 50
ratio = 20/3
radius = thickness = 20
velx = 15
vely=0
elastic_coeff = 0.85
g =1# 980*ratio

run = True
time = 30
p = False
x1 = y1 = 0
while run:
    pygame.time.delay(time)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    
    win.fill((255,255,255))

    if pygame.mouse.get_pressed()[0] and abs(x1-x)<25 and abs(y1-y)<25:
        if  pygame.mouse.get_pressed()[2] == False and p == False:
            x1,y1 = pygame.mouse.get_pos()
            x,y = pygame.mouse.get_pos()
            velx = vely = 0
        elif pygame.mouse.get_pressed()[2]:
            x2,y2 =pygame.mouse.get_pos()
            pygame.draw.circle(win, (255,0,0), (x2, y2), radius, 2) 
            p = True
        elif p and pygame.mouse.get_pressed()[2] == False:
            pygame.draw.circle(win, (255,0,0), (x2, y2), radius, 2)
            velx = (x2-x)/5
            vely = (y2-y)/5
        
    else:
        ytemp = y+vely+g/2
        if ymin+2*radius<=ytemp+radius <= ymax:
            vely = vely +g
            y = y+vely+g/2
        elif ymin+2*radius>ytemp+radius:
            vely = - elastic_coeff*vely
            y = ymin+radius
        elif ytemp+radius > ymax:
            vely = - elastic_coeff*vely
            y = ymax-radius    


        xtemp = x+velx

        if xmin+2*radius<=xtemp+radius <= xmax and y !=ymax-radius :
            x = x+velx
        elif xmin+2*radius<=xtemp+radius <= xmax and y==ymax-radius :
            velx = 0.7*velx
            x = x+velx
        elif xmin+2*radius>xtemp+radius:
            velx = - elastic_coeff*velx
            x = xmin+radius
        elif xtemp+radius > xmax:
            velx = - elastic_coeff*velx
            x = xmax-radius
        x1,y1 = pygame.mouse.get_pos()
        p = False
    pygame.draw.circle(win, (0,0,0), (x, y), radius, thickness)  
    pygame.display.update() 
    
pygame.quit()
```
